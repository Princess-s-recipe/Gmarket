```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver

def crawl_pages(url_template):
    chrome_driver_path = "./driver"
    driver = webdriver.Chrome(chrome_driver_path)

    page = 1
    while True:
        url = url_template.format(page=page)
        driver.get(url)
        
        # 페이지의 끝까지 스크롤하는 작업을 반복
        while True:
            last_height = driver.execute_script("return document.body.scrollHeight")  # 현재 위치 저장
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # 스크롤을 끝까지 내림
            time.sleep(1)  # 페이지 로딩 기다림
            new_height = driver.execute_script("return document.body.scrollHeight")
            
            # 새로운 높이와 이전 높이가 같아진다면, 더 이상 스크롤할 수 없으므로 반복문을 종료
            if new_height == last_height:
                break
                
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        
        # Titles
        titles = soup.select("div.box__item-title")
        
        # Sale_Prices
        sale_prices = soup.select("div.box__price-seller")
        
        for title, price in zip(titles, sale_prices):
            print(f"{title.text} - {price.text}")
        
        # 다음 버튼이 있는지 확인하고 없으면 마지막 페이지이니 break
        next_button = soup.select_one("a.link__page-next")
        
        if not next_button:
            break
            
        page += 1

    #driver.quit()

# Example usage
#url_template = 'https://browse.gmarket.co.kr/search?keyword=%EB%A7%A5%EB%B6%81&t=a&k=0&p={page}'
# 페이지가 너무 많아서 마지막 페이지까지 잘 동작하는지 확인하기 위한 스마일배송 맥북으로 설정
url_template = 'https://browse.gmarket.co.kr/search?keyword=%EB%A7%A5%EB%B6%81&t=s&k=0&p={page}'
crawl_pages(url_template)
```
